```python
dfs = [apple, chase, microsoft, tesla]
```

```python
apple.columns
```

for df in dfs:
    df.columns = df.columns.str.strip()

```python
for df in dfs:
    df['MA5'] = df.Close.rolling(5).mean()
    df['MA20'] = df.Close.rolling(20).mean()
    df['MA50'] = df.Close.rolling(50).mean()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day close price'] = df['Close'].shift(1)
```

```python
for df in dfs:
    df['Change in price'] = df['Close']- df['Previous day close price']
```

```python
for df in dfs:
    df['Percent change in price'] = df.Close.pct_change()
```

```python
apple.head()
```

```python
for df in dfs:
    df['Previous day volume'] = df['Volume'].shift(1)
```

```python
for df in dfs:
    df['Change in volume'] = df['Volume']- df['Previous day volume']    
```

```python
for df in dfs:
    df['Percent change in volume'] = df.Volume.pct_change()
```

```python
for df in dfs:
    df['VMA5'] = df.Volume.rolling(5).mean()
    df['VMA20'] = df.Volume.rolling(20).mean()
```

```python
for df in dfs:
    df['Volume_Osc'] = (df['VMA5']- df['VMA20'])/df['VMA20'] * 100
```

```python
for df in dfs:
    df['Open'] = pd.to_numeric(df['Open'], errors='coerce')
```

```python
apple.dtypes
```

```python
for df in dfs:
    df['Previous day open price'] = df['Open'].shift(1)
for df in dfs:
    df['Change in open price'] = df['Open']- df['Previous day open price']
for df in dfs:
    df['Percent change in open price'] = df.Open.pct_change()
```

```python
for df in dfs:
    df['Previous day high price'] = df['High'].shift(1)
for df in dfs:
    df['Change in high price'] = df['High']- df['Previous day high price']
for df in dfs:
    df['Percent change in high price'] = df.High.pct_change()
```

```python
for df in dfs:
    df['Previous day low price'] = df['Low'].shift(1)
for df in dfs:
    df['Change in low price'] = df['Low']- df['Previous day low price']
for df in dfs:
    df['Percent change in low price'] = df.Low.pct_change()
```
